```python
def add_contact():
    
    def get_data(list_data, string_data):
        while True:
            d = input(f'Enter {string_data} {len(list_data)+1} : ')
            list_data.append(d)

            if input('Do you wish to add more ?(Y/N): ').upper() != 'Y':
                break 
    
    name = input('Enter name: ')
    if name in contacts:
        print(f'{name} already exist')
    else:
        add = input('Enter address: ')
        phones = []
        emails = []

        get_data(phones, 'phone no')
        get_data(emails, 'email')

        contacts[name] = {'add': add, 'phones': phones, 'emails': emails}
        print(contacts)
        
def search_contact(name):
    if name in contacts:
        print(f'Name: {name}')
        print(f"Address: {contacts[name]['add']}")
        print(f"Phone Nos: {contacts[name]['phones']}")
        print(f"Emails: {contacts[name]['emails']}")
        return True
    else:
        return False
                

def edit_contact():    
    
    def process_data(data):
        
        def edit(option):
            for i in range(len(contacts[name][data])):
                print(f"{i}: {contacts[name][data][i]}")

            index = int(input('Enter option: '))
            if option == 'edit':
                new_data = input(f'Enter New {data}: ')
                contacts[name][data][index] = new_data
                print(f'{new_data} successfully updated.') 
            elif option == 'del':
                temp = contacts[name][data][index]
                del contacts[name][data][index]
                print(f'{temp} successfully deleted.') 
            
       
        while True:
            print('[1]: Add New\n[2]: Edit\n[3]: Delete\n[#]: Previous Menu')
            option = input('Enter option: ')

            if option == '1':
                new_data = input(f'Enter New {data} No: ')
                contacts[name][data].append(new_data)
                print(f'{new_data} successfully added.')
            elif option == '2':
                edit('edit')
            elif option == '3':
                edit('del')
            elif option == '#':
                break

    
    name = input('Enter name: ')
    if search_contact(name):
        while True:
            print('[1]: Edit Address\n[2]: Edit Phone Nos\n[3]: Edit Emails\n[#]: Previous Menu')
            option = input('Enter option: ')

            if option == '1':
                old_add = contacts[name]['add']
                add = input('Enter New Address: ')
                contacts[name]['add'] = add
                print(f'{old_add} successfully updated to {add}')
            elif option == '2':
                process_data('phones')
            elif option == '3':
                process_data('emails')
            elif option == '#':
                break  
    else:
        print(f'{name} does not exist!')

        
def delete_contact():
    name = input('Enter name: ')
    if name in contacts:
        del contacts[name]
        print(f'{name} successfully deleted!')
    else:
        print(f'{name} does not exist!')
        

        
contacts = {}

def app():
    trial = 0
    while True:  
        print('[1]: Add Contact\n[2]: Edit Contact\n',
            '[3]: Delete Contact\n[4]: Search Contact\n[#]: Exit', sep='')

        option = input('Enter option: ')

        if option == '1':
            add_contact()
        elif option == '2':
            edit_contact()
        elif option == '3':
            delete_contact()
        elif option == '4':
            name = input('Enter name: ')
            if search_contact(name):
                pass
            else:
                print(f'{name} does not exist!')
        elif option == '#':
            print('Goodbye!!!')
            break
        else:
            print('Invalid option')
            trial += 1

            if trial == 3:
                print('Technical Issue')
                break
                
                
                
if __name__ == "__main__":
    app()
```
